```python
import numpy as np
import pandas as pd
tour_comp = pd.read_csv("TourneyCompactResults.csv");
seeds = pd.read_csv("TourneySeeds.csv");
```

```python
#seeds['region'] = seeds['Seed'].astype(str).str[0:1];
#seeds["Seed"].replace(seeds["Seed"][0:1],"_");
seeds["Seed"]= [i[1:3] for i in seeds["Seed"]]
#seeds['Seed']= seeds['Seed'].str[0:3].astype(int)
seeds.head()
```

```python
tour_comp = tour_comp.rename(columns={"Wteam":"team1", "Lteam":"team2"});
tour_comp["result"] = np.repeat(1,2050);
for i in range(2050):
    if tour_comp.team1[i] > tour_comp.team2[i]:       
        team1 = tour_comp.team2[i]
        team2 = tour_comp.team1[i]
        tour_comp.set_value(i, "team1", team1)
        tour_comp.set_value(i, "team2", team2)
        tour_comp.set_value(i, "result", 0)
```

```python
tour_comp.tail()
```

```python
tour_comp=tour_comp[tour_comp['Season']>2002];
tour_comp=tour_comp[tour_comp['Season']<2017];
seeds=seeds[seeds['Season']>2002];
seeds=seeds[seeds['Season']<2017];
```

```python
data = tour_comp[tour_comp.columns[[0,1,2,4,8]] ];
data = pd.merge(data, seeds, how='left', left_on=["Season","team1"], right_on=["Season","Team"])
data = pd.merge(data, seeds, how='left', left_on=["Season","team2"], right_on=["Season","Team"])
```

```python
data['Seed_x']= data.Seed_x.astype('int');
data['Seed_y']= data.Seed_y.astype('int');
data['different_seed']=data['Seed_x'].values-data['Seed_y'].values;
data.head()
```

```python
#data=data[data.columns[[0,1,2,3,5,7,8,10,11,4]]];
data=data[data.columns[[0,1,2,3,4,5,7,9]]]
data.head()
```

```python
regular_score=pd.read_csv('regular_score.csv');
detail=pd.read_csv('detail.csv');
del regular_score['Unnamed: 0'];
regular_team1=regular_score.rename(columns={'team':'team1'})
#regular_team1=regular_team1.rename(columns={'score':'score1'})
regular_team1.head();
regular_team2=regular_score.rename(columns={'team':'team2'})
#regular_team2=regular_team2.rename(columns={'score':'score2'})
regular_team2.head();
detail1=detail.rename(columns={'team':'team1'})
detail1.head();
detail2=detail.rename(columns={'team':'team2'})
regular_team1.head()
```
