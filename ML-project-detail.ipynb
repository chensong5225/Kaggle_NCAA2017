```python
import pandas as pd
from pandas import read_csv
import numpy as np
import types
import csv
detail = pd.read_csv("RegularSeasonDetailedResults.csv");
detail.head()
```

## mean of detail for each team in each Seaon

```python

selected_columns1 = [col for col in detail.columns if "W" in col];
selected_columns1.append('Season');
selected_columns1.append('Daynum');
win_detail = detail[selected_columns1];
selected_columns2 = [col for col in detail.columns if "L" in col];
selected_columns2.append('Season');
selected_columns2.append('Daynum');
lose_detail = detail[selected_columns2];
win_detail.columns = win_detail.columns.str.replace("W", "")
lose_detail.columns=lose_detail.columns.str.replace("L","");
detail = [win_detail, lose_detail]
detail = pd.concat(detail);
```

```python
detail_mean=detail.groupby(['team','Season'])['fgm', 'fga', 'fgm3', 'fga3', 'ftm', 'fta',
       'or', 'dr', 'ast', 'to', 'stl', 'blk', 'pf'].mean();
detail_mean=detail_mean.reset_index()
```

```python
detail_mean.to_csv('detail.csv')
```
